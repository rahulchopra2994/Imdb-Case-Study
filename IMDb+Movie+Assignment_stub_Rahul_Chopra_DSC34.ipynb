3. First group the dataframe `df_by_genre` by `genre_1` and find the sum of all the numeric columns such as `cnt`, columns related to CVotes and Votes columns and store it in a dataframe `df_by_g1`.
4. Perform the same operation for `genre_2` and `genre_3` and store it dataframes `df_by_g2` and `df_by_g3` respectively.

For this, add the three dataframes and store it in a new dataframe `df_add`, so that the corresponding values of Votes/CVotes get added for each genre.There is a function called `add()` in pandas which lets you do this. You can refer to this link to see how this function works. https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.add.html

7. Now, take the mean of all the numeric columns by dividing them with the column value `cnt` and store it back to the same dataframe. We will be using this dataframe for further analysis in this task unless it is explicitly mentioned to use the dataframe `movies`.
8. Since the number of votes can't be a fraction, type cast all the CVotes related columns to integers. Also, round off all the Votes related columns upto two digits after the decimal point.


```python
# Create the dataframe df_by_genre
movies.columns
```

```python
df_by_genre=movies[['genre_1', 'genre_2', 'genre_3', 'CVotes10',
       'CVotes09', 'CVotes08', 'CVotes07', 'CVotes06', 'CVotes05', 'CVotes04',
       'CVotes03', 'CVotes02', 'CVotes01', 'CVotesMale', 'CVotesFemale',
       'CVotesU18', 'CVotesU18M', 'CVotesU18F', 'CVotes1829', 'CVotes1829M',
       'CVotes1829F', 'CVotes3044', 'CVotes3044M', 'CVotes3044F', 'CVotes45A',
       'CVotes45AM', 'CVotes45AF', 'CVotes1000', 'CVotesUS', 'CVotesnUS',
       'VotesM', 'VotesF', 'VotesU18', 'VotesU18M', 'VotesU18F', 'Votes1829',
       'Votes1829M', 'Votes1829F', 'Votes3044', 'Votes3044M', 'Votes3044F',
       'Votes45A', 'Votes45AM', 'Votes45AF', 'Votes1000', 'VotesUS',
       'VotesnUS']]
```

```python
# Create a column cnt and initialize it to 1
df_by_genre['cnt'] = 1
df_by_genre.head()
```

```python
# Group the movies by individual genres
df_by_g1 = df_by_genre.groupby(by="genre_1").sum()
```

```python
df_by_g1.head()
```

```python
df_by_g2 = df_by_genre.groupby(by="genre_2").sum()
```

```python
df_by_g2.head()
```

```python
df_by_g3 = df_by_genre.groupby(by="genre_3").sum()
```

```python
df_by_g3.head()
```

```python
# Add the grouped data frames and store it in a new data frame
df1=df_by_g1.add(df_by_g2,fill_value=0)
df_add = df1.add(df_by_g3,fill_value=0)
df_add.head()
```

```python
# Extract genres with atleast 10 occurences
genre_top10 = df_add[df_add.cnt>=10]
genre_top10.head()
```

```python
genre_top10.shape
```

```python
# Take the mean for every column by dividing with cnt 
genre_top10.loc[:,'CVotes10':'VotesnUS'] = genre_top10.loc[:,'CVotes10':'VotesnUS'].divide(genre_top10.cnt, axis=0)
```

Similarly, make another subplot that shows how non US voters have voted for the US and non-US movies by plotting `CVotesnUS` for both the US and non-US movies. 3. Again do a similar analysis but with the ratings. Make a boxplot that shows how the ratings from the US people i.e. `VotesUS` is varying for the US and non-US movies. Similarly, make another subplot that shows how `VotesnUS` is varying for the US and non-US movies. Write any of your two inferences/observations from these plots.

Note : Use `movies` dataframe for this subtask. Make use of this documention to format your boxplot - https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
# Creating IFUS column
movies['IFUS'] = movies['Country'].apply(lambda x: x if x=='USA' else 'non-USA')
movies.head()
```

```python
# Box plot - 1: CVotesUS(y) vs IFUS(x)
plt.figure(figsize=(21,8))
plt.subplot(1,2,1)
sns.boxplot(x=movies.IFUS,y=movies.CVotesUS)
plt.xlabel('Movie Nationality',fontsize = '15',fontweight='bold')
plt.ylabel('US Votes',fontsize = '15',fontweight='bold')
plt.title('Movie Nationality vs US Votes', fontsize = '18', fontweight = 'bold')
plt.subplot(1,2,2)
sns.boxplot(x=movies.IFUS,y=movies.CVotesnUS)
plt.xlabel('Movie Nationality',fontsize = '15',fontweight='bold')
plt.ylabel('Non-US Votes',fontsize = '15',fontweight='bold')
plt.title('Movie Nationality vs Non-US Votes', fontsize = '18', fontweight = 'bold')
plt.subplot(1,2,2)
plt.show()
```

**`Inferences:`** Write your two inferences/observations below:
- Inference 1: Firstly, it can be clearly seen that the interquartile range is spread more largely for non USA movies for both USA and non-USA voters, however the median and upper fence seems to be larger for the USA movies in both voter categories. With this the inference we can draw is that American movies tend to have a higher number of votes casted by voters from all nationalities as compared to non-USA movies.      
- Inference 2: There are a few outliers for movies created in USA from both USA and non-USA voters suggesting that some American movies are highly popular for voters irrespective of the nationality of the voters.

```python
# Box plot - 2: VotesUS(y) vs IFUS(x)

plt.figure(figsize=(21,8))
plt.subplot(1,2,1)
sns.boxplot(x=movies.IFUS,y=movies.VotesUS)
plt.xlabel('Movie Nationality',fontsize = '15',fontweight='bold')
plt.ylabel('US Average Ratings',fontsize = '15',fontweight='bold')
plt.title('Movie Nationality vs US Ratings', fontsize = '18', fontweight = 'bold')
plt.xticks(fontsize = '13',fontweight='bold')
plt.subplot(1,2,2)
sns.boxplot(x=movies.IFUS,y=movies.VotesnUS)
plt.xlabel('Movie Nationality',fontsize = '15',fontweight='bold')
plt.ylabel('Non-US Ratings',fontsize = '15',fontweight='bold')
plt.title('Movie Nationality vs Non-US Ratings', fontsize = '18', fontweight = 'bold')
plt.xticks(fontsize = '13',fontweight='bold')
plt.subplot(1,2,2)
plt.show()
```

**`Inferences:`** Write your two inferences/observations below:
- Inference 1: In the case of US based Ratings the IQR for USA movies is from rating of 7.8 to 8.1, whereas for non-USA movies it is only 7.8-8. Whereas, in the case of non-US based Ratings, the IQR for USA movies is 7.6-7.9,while its 7.6- 8 almost in the case of non USA movies. This suggests that people tend to rate higher based on commonalities in the Movie's country of origin and their own nationality.
- Inference 2: For both US and non-US based Ratings, there are outliers for USA movies that suggest some American films are rated highly(more than 8.4)
